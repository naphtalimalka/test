```python
from PIL import Image
import os
from google.colab import drive

drive.mount('/content/drive', force_remount=True)
dossier_drive = "/content/drive/MyDrive/meching/Input/"
fichiers = ["face.png", "droite.png", "dos.png", "gauche.png", "haut.png", "bas.png"]

images = [Image.open(os.path.join(dossier_drive, f)).convert("RGB").resize((512, 512)) for f in fichiers]
grille = Image.new('RGB', (512 * 3, 512 * 2))

for i, img in enumerate(images):
    grille.paste(img, (i % 3 * 512, i // 3 * 512))

grille.save("/content/LGM/input_grid.png")
print("✅ Grille créée dans /content/LGM/input_grid.png")
```
