```python
import os
import glob

# 1. On cherche où sont les fichiers JPG dans tout le dossier cloné
photos = glob.glob('/content/dataset_buddha/**/*.jpg', recursive=True)

if not photos:
    # Si pas de JPG, on cherche les PNG
    photos = glob.glob('/content/dataset_buddha/**/*.png', recursive=True)

if photos:
    # On récupère le dossier qui contient la première photo trouvée
    input_dir = os.path.dirname(photos[0])
    print(f"✅ Photos trouvées dans : {input_dir}")

    # Configuration des autres dossiers
    output_dir = "/content/output_buddha"
    cache_dir = "/content/cache"
    meshroom_path = "/content/meshroom/Meshroom-2019.1.0/meshroom_photogrammetry"

    !mkdir -p "{output_dir}"
    !mkdir -p "{cache_dir}"

    # Lancement de Meshroom
    print("🚀 Démarrage de la reconstruction...")
    !{meshroom_path} --input "{input_dir}" --output "{output_dir}" --cache "{cache_dir}"
else:
    print("❌ Erreur : Aucune photo trouvée dans /content/dataset_buddha/")
    print("Vérifie le contenu du dossier avec la commande : !ls -R /content/dataset_buddha")
```
