```python
from PIL import Image
import os
from google.colab import drive

# Montage du drive
drive.mount('/content/drive', force_remount=True)

dossier_drive = "/content/drive/MyDrive/meching/Input/"
fichiers = ["face.png", "droite.png", "dos.png", "gauche.png", "haut.png", "bas.png"]

try:
    images = [Image.open(os.path.join(dossier_drive, f)).convert("RGB").resize((512, 512)) for f in fichiers]
    grille = Image.new('RGB', (512 * 3, 512 * 2))

    for i, img in enumerate(images):
        grille.paste(img, (i % 3 * 512, i // 3 * 512))

    grille.save("/content/LGM/input_grid.png")
    print("✅ Grille 'input_grid.png' prête dans le dossier LGM !")
except Exception as e:
    print(f"❌ Erreur : {e}. Vérifie tes noms de fichiers dans Drive.")
```

```python
import shutil
import os

# Dossier destination sur ton Drive
dest_drive = "/content/drive/MyDrive/meching/Output/"
os.makedirs(dest_drive, exist_ok=True)

# Dossier où LGM enregistre les exports
source_folder = "/content/LGM/outputs/"

if os.path.exists(source_folder) and os.listdir(source_folder):
    for item in os.listdir(source_folder):
        s = os.path.join(source_folder, item)
        d = os.path.join(dest_drive, item)
        if os.path.isdir(s):
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d)
    print(f"✅ Ton monstre 3D est sauvegardé ici : {dest_drive}")
else:
    print("❌ Aucun fichier trouvé. As-tu bien cliqué sur 'Export' dans l'interface ?")
```
