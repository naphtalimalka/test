```python
from PIL import Image
import os

# --- CONFIGURATION ---
# 1. Copie ici le chemin du dossier où sont tes 6 images
dossier = "/content/drive/MyDrive/TON_DOSSIER_IMAGES/"

# 2. Liste tes fichiers dans cet ORDRE PRÉCIS
fichiers = [
    "face.png",    # Image 1
    "droite.png",  # Image 2
    "dos.png",     # Image 3
    "gauche.png",  # Image 4
    "haut.png",    # Image 5
    "bas.png"      # Image 6
]

# --- CRÉATION DE LA GRILLE ---
images = [Image.open(os.path.join(dossier, f)).convert("RGB").resize((512, 512)) for f in fichiers]
w, h = 512, 512
grille = Image.new('RGB', (w * 3, h * 2))

# Placement : Face, Droite, Dos (ligne 1) / Gauche, Haut, Bas (ligne 2)
for i, img in enumerate(images):
    grille.paste(img, (i % 3 * w, i // 3 * h))

grille.save("/content/ma_grille_6_vues.png")
print("✅ Ta grille est prête ! Elle se trouve ici : /content/ma_grille_6_vues.png")
```
