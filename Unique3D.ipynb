```python
from PIL import Image
import os

# --- CONFIGURATION ---
dossier_drive = "/content/drive/MyDrive/meching/Input/"
fichiers = ["face.png", "droite.png", "dos.png", "gauche.png", "haut.png", "bas.png"]
# ---------------------

try:
    images = [Image.open(os.path.join(dossier_drive, f)).convert("RGB").resize((512, 512)) for f in fichiers]
    w, h = 512, 512
    grille = Image.new('RGB', (w * 3, h * 2))

    for i, img in enumerate(images):
        grille.paste(img, (i % 3 * w, i // 3 * h))

    # Sauvegarde locale pour l'IA
    grille.save("/content/Unique3D_input.png")
    # Sauvegarde de sécurité sur ton Drive
    grille.save(dossier_drive + "grille_generee_unique3d.png")
    print("✅ Grille prête et sauvegardée sur le Drive !")
except Exception as e:
    print(f"❌ Erreur : {e}. Vérifie tes noms de fichiers.")
```
