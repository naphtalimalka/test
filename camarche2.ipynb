```python
import os
import glob

# 1. On cherche les photos avec TOUTES les extensions possibles
path_to_search = '/content/dataset_buddha'
photos = []
for ext in ['*.jpg', '*.JPG', '*.jpeg', '*.JPEG', '*.png', '*.PNG']:
    photos.extend(glob.glob(f'{path_to_search}/**/{ext}', recursive=True))

if photos:
    # On trie pour avoir un ordre logique
    photos.sort()
    input_dir = os.path.dirname(photos[0])
    print(f"✅ SUCCÈS ! {len(photos)} photos trouvées dans : {input_dir}")

    # CONFIGURATION
    output_dir = "/content/output_buddha"
    cache_dir = "/content/cache"
    meshroom_path = "/content/meshroom/meshroom_photogrammetry"

    !mkdir -p "{output_dir}"
    !mkdir -p "{cache_dir}"
    !chmod +x "{meshroom_path}"

    # 2. LANCEMENT DU CALCUL
    print("🚀 Démarrage de la reconstruction (étape 1/13)...")
    !{meshroom_path} --input "{input_dir}" --output "{output_dir}" --cache "{cache_dir}"
else:
    print("❌ Toujours rien. Tentative de diagnostic forcé :")
    !find /content/dataset_buddha -maxdepth 3
```
